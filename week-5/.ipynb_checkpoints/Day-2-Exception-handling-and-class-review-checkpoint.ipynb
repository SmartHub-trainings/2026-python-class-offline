```python
try:
    a= float(input("Enter a number: "))
    b= float(input("Enter a number: "))
    c= a/b
    print(c)
except ZeroDivisionError:
    print("You can't divide by 0.")
except ValueError:
    print("Only numbers are allowed")
except:
    print("You did something wrong!")
```

```python
try:
    a= float(input("Enter a number: "))
    b= float(input("Enter a number: "))
    c= a/b
    print(c)
except ZeroDivisionError:
    print("You can't divide by 0.")
except ValueError:
    print("Only numbers are allowed")
except:
    print("You did something wrong!")
else:
    print("Well, no error occured")
finally:
    print("Program done")
```

```python
menu="""1.Fride rice
2.Rice and Stew
3.Beans
4.Egusi Soup
5.Afang Soup
6.Pepper Soup
7.Swallow"""
print(menu)
choice =int(input('What do you want?(1-7)'))
price={1:5000,2:2500,
       3:1000,4:1700,
      5:2000,
      6:2500,
      "7.1":700,
      "7.2":500}
if choice ==7:
    swallow_menu ="1.Garri\n2.Loi"
    print(swallow_menu)
    swallow_choice =int(input('What do you want?(1 or 2)'))
    choice = f"{choice}.{swallow_choice}"
print(price[choice])
```

```python
total_cost =0
while True:
    menu="""1.Fride rice
2.Rice and Stew
3.Beans
4.Egusi Soup
5.Afang Soup
6.Pepper Soup
7.Swallow"""
    print(menu)
    choice =int(input('What do you want?(1-7)'))
    price={1:5000,2:2500,
           3:1000,4:1700,
          5:2000,
          6:2500,
          "7.1":700,
          "7.2":500}
    if choice ==7:
        swallow_menu ="1.Garri\n2.Loi"
        print(swallow_menu)
        swallow_choice =int(input('What do you want?(1 or 2)'))
        choice = f"{choice}.{swallow_choice}"
    print(price[choice])
    total_cost = total_cost+price[choice]
    need_something_else = input("Do you care for something else? (y/n): ")
    if need_something_else.lower() in ["n","no","nah","nope"]:
        break
print(f"Total Cost is {total_cost}")
    
    
```

```python
total_cost =0
while True:
    menu="""1.Fride rice
2.Rice and Stew
3.Beans
4.Egusi Soup
5.Afang Soup
6.Pepper Soup
7.Swallow"""
    print(menu)
    choice =int(input('What do you want?(1-7)'))
    price={1:5000,2:2500,
           3:1000,4:1700,
          5:2000,
          6:2500,
          "7.1":700,
          "7.2":500}
    if choice ==7:
        swallow_menu ="1.Garri\n2.Loi"
        print(swallow_menu)
        swallow_choice =int(input('What do you want?(1 or 2)'))
        choice = f"{choice}.{swallow_choice}"
    print(price[choice])
    total_cost = total_cost+price[choice]
    print(f"Total cost {total_cost}\n")
    
    need_something_else = input("Do you care for something else? (y/n): ")
    if need_something_else.lower() in ["n","no","nah","nope"]:
        break
print(f"Total Cost is {total_cost}")
    
    
```
