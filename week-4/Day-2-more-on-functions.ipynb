 Create a function that computes simple interest and amount when principal, rate and time are given


```python
def get_returns(p,r,t):
    si= p*r*t/100
    amount = p+si
    print(f"Simple interest is {si}")
    print(f"Your amount at the end of {t} yrs is {amount}")
```

```python
get_returns(1e4,5,2)
```

```python
get_returns(5e6,10,4)
```
