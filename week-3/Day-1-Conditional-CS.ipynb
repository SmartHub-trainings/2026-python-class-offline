```python
price_of_rice = 900
price_of_chicken=1600
num_of_plates_of_rice = int(input("Enter number of plates of rice: "))
num_of_pieces_of_chicken = int(input("Enter number of  pieces of chicken: "))
total= price_of_chicken*num_of_pieces_of_chicken+price_of_rice*num_of_plates_of_rice
print(f"Your total bill is #{total}")
if total>=5000 and total <10000:
    disc_perc=0.05
elif total >=10000:
    disc_perc=0.1
else:
    disc_perc=0
discount = disc_perc*total
print(f"Discount :{discount}")
total = total -discount
print(f"You have a discount of  #{discount}")
print(f"You'r to pay #{total}")
    
```
