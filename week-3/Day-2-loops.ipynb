```python
for i in range(6):
    name = input("Enter your name: ")
    if not name.startswith("R") and not name.startswith("F"):
        print(f"Hello, {name}")
```

```python
for i in range(6):
    name = input("Enter your name: ").title()
    if not name.startswith("R") and not name.startswith("F"):
        print(f"Hello, {name}")
```

```python
for i in range(6):
    name = input("Enter your name: ").title()
    if not (name.startswith("R") or  name.startswith("F")):
        print(f"Hello, {name}")
```

## Break and Continue statements 

```python
for i in range(10):
    if i ==5:
        continue
    else:
        print(i)
```

```python
for i in range(10):
    if i ==5:
        break
    else:
        print(i)
```

```python
for i in range(10):
    if i ==5:
        continue
    print(i)
```
